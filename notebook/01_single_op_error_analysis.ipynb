```python
from src.single_operation import SingleOperationExperiment
from src.core.plotter import plot_error_heatmap, plot_error_distribution
import numpy as np
```

```python

# Configuration parameters
sizes = range(1, 1000)  # Varying accumulation lengths
repeats = 100  # Number of random trials
value_range = (0, 2)  # Range of input values
dtype = np.float32  # Precision to test (e.g., np.float16, np.float32)

# Initialize and run experiment
exp = SingleOperationExperiment(
    dtype_list=[dtype],
    sizes=sizes,
    repeats=repeats,
    value_range=value_range
)
exp.run()
# Extract results and visualize
errors_matrix = np.array(exp.results[dtype])
plot_error_heatmap(errors_matrix, title="Heatmap of Float32 Error")
plot_error_distribution(errors_matrix, title="Distribution of Float32 Error")
```

```python
# Configuration parameters
sizes = range(1, 1000)  # Varying accumulation lengths
repeats = 100  # Number of random trials
value_range = (-1, 1)  # Range of input values
dtype = np.float32  # Precision to test (e.g., np.float16, np.float32)

# Initialize and run experiment
exp = SingleOperationExperiment(
    dtype_list=[dtype],
    sizes=sizes,
    repeats=repeats,
    value_range=value_range
)
exp.run()
# Extract results and visualize
errors_matrix = np.array(exp.results[dtype])
plot_error_heatmap(errors_matrix, title="Heatmap of Float32 Error", ylim=(-6e-8, 6e-8))
plot_error_distribution(errors_matrix, title="Distribution of Float32 Error", ylim=(-6e-8, 6e-8))
```
